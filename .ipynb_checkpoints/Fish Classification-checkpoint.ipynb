```python
directory = './Dataset/'

# Create the training dataset
train_dataset_manual = tf.keras.preprocessing.image_dataset_from_directory(
    directory=directory,
    labels='inferred',
    label_mode='int',
    class_names=None,
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=123,
    validation_split=None,
    subset=None,
    interpolation='bilinear',
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
def apply_otsu_watershed_batch(img_batch):
    processed = []

    for img in img_batch:
        img = img.astype(np.uint8)

        # Convert to grayscale
        if len(img.shape) == 2:
            gray = img
        elif len(img.shape) == 3 and img.shape[2] == 1:
            gray = np.squeeze(img, axis=2)
        elif len(img.shape) == 3 and img.shape[2] == 3:
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        else:
            raise ValueError(f"Unexpected image shape: {img.shape}")

        # Otsu's threshold
        _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Morphological opening (remove noise)
        kernel = np.ones((3, 3), np.uint8)
        opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)

        # Sure background by dilation
        sure_bg = cv2.dilate(opening, kernel, iterations=3)

        # Sure foreground by distance transform
        dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
        _, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)
        sure_fg = np.uint8(sure_fg)

        # Unknown region
        unknown = cv2.subtract(sure_bg, sure_fg)

        # Marker labelling
        _, markers = cv2.connectedComponents(sure_fg)
        markers = markers + 1
        markers[unknown == 255] = 0

        # Convert grayscale or RGB to 3-channel BGR for watershed
        if len(img.shape) == 2:
            img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        else:
            img_color = img.copy()

        # Watershed
        markers = cv2.watershed(img_color, markers)
        img_color[markers == -1] = [255, 0, 0]  # Mark boundaries in red

        # Convert to RGB (TensorFlow prefers RGB over BGR)
        img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

        processed.append(img_rgb)

    return np.array(processed, dtype=np.uint8)

def otsu_watershed_preprocess(image, label):
    image = tf.numpy_function(apply_otsu_watershed_batch, [image], tf.uint8)
    image.set_shape([None, 256, 256, 3])  # maintain batch image shape
    return image, label
```

```python
train_dataset_manual = train_dataset_manual.map(otsu_watershed_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
```

DATA PROCESSING

```python
IM_SIZE = 128
def resize_rescale(image,label):
    image = tf.image.resize(image , (IM_SIZE,IM_SIZE))
    image = image /255.0 
    return image,label
```

```python
train_dataset = train_dataset.map(resize_rescale)
val_dataset = val_dataset.map(resize_rescale)
```

```python
for image,label in train_dataset.take(1):
    print(image , label)
```

```python
# Load pre-trained RESNET (without top layers)
base_model = ResNet152V2(
    weights='imagenet',
    include_top=False,
    input_shape=(128, 128, 3))
base_model.trainable = False  # Freeze base layers


model_res = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu',kernel_regularizer=l2(0.01)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(1, activation='sigmoid')
])
model_res.summary()
```

```python
metrics = [TruePositives(name = 'tp'), FalsePositives(name = 'fp') , TrueNegatives(name = 'tn') , FalseNegatives(name = 'fn') , 
          Precision(name = 'Precision') , Recall(name='recall'),AUC(name='auc'),BinaryAccuracy(name='accuracy')]
```

```python
model_res.compile(optimizer=Adam(0.00001),
              loss = BinaryCrossentropy(),
              metrics = metrics)
```

```python
es = EarlyStopping(monitor='val_accuracy',
    min_delta=0.0001,
    patience=3,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True)
lr = ReduceLROnPlateau(monitor='val_loss',
    factor=0.5,
    patience=1,
    verbose=0,
    mode='auto',
    min_delta=0.0001,
    cooldown=0,
    min_lr=0)
```

```python
history_res = model_res.fit(train_dataset , validation_data=val_dataset ,epochs = 40 ,verbose =1,callbacks=[es,lr])
```

```python
test_dataset = test_dataset.map(resize_rescale)
```

```python
# Initialize lists to store predictions and true labels
all_true_labels = []
all_predicted_probs = []

# Process test data in small batches
for batch_images, batch_labels in tqdm(test_dataset):
    # Predict probabilities for the current batch (output is [batch_size, 1] for binary classification)
    batch_preds = model_res.predict(batch_images, verbose=0)
    all_predicted_probs.append(batch_preds)
    all_true_labels.append(batch_labels.numpy())

# Concatenate all batches
y_true = np.concatenate(all_true_labels, axis=0)
y_pred_probs = np.concatenate(all_predicted_probs, axis=0)

# Convert probabilities to binary predictions (0 or 1) using threshold=0.5
y_pred = (y_pred_probs > 0.5).astype(int)

# Compute metrics
test_accuracy = accuracy_score(y_true, y_pred)
test_precision = precision_score(y_true, y_pred)
test_recall = recall_score(y_true, y_pred)
test_f1 = f1_score(y_true, y_pred)
test_roc_auc = roc_auc_score(y_true, y_pred_probs)  # Requires probabilities, not binary predictions

print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")
print(f"Test F1-Score: {test_f1:.4f}")
print(f"Test ROC-AUC: {test_roc_auc:.4f}")

# Detailed classification report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=["Class 0", "Class 1"]))
```
